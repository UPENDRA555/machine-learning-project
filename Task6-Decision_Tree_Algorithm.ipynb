### Import Required Librearies

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

#### Import Datasets

```python
data=pd.read_csv('D:\DATASETS\IRISDATASETS\Iris.csv')
data.head()
```

```python
 data.shape
```

```python
data['Species'].value_counts()
```

```python
data[data['Species']=='Iris-versicolor'].head()
```

```python
data['Species']=data['Species'].replace(['Iris-setosa', 'Iris-versicolor', 'Iris-virginica'], [0, 1, 2])
data.head()
```

```python
X=data.drop(['Id','Species'], axis=1)
Y=pd.DataFrame(data['Species'])
```

```python
X=np.array(X)
X
```

```python
Y=np.array(Y)
Y
```

```python
from sklearn.tree import DecisionTreeClassifier
classifier=DecisionTreeClassifier(random_state=0)
classifier.fit(X,Y)
```

```python
from sklearn.tree import plot_tree
plt.figure(figsize=(48,18), dpi=150)
plot_tree(classifier, filled=True)
plt.show()
```
